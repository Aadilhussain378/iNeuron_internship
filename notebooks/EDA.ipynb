## Travel Package Purchase

## Customer Details
- CustomerID: A unique identifier assigned to each customer. This helps distinguish and track individual customers in the dataset.

- ProdTaken: A binary variable indicating whether the customer decided to purchase the travel product (1) or not (0). This is a crucial target variable for predictive modeling.

- Age: Represents the age of the customer, providing demographic information that could influence travel preferences and decision-making.

- TypeofContact: Describes the method or type of contact made with the customer (e.g., email, phone call, in-person meeting). This information is relevant for understanding communication channels.

- CityTier: Indicates the city tier, classifying the city based on factors like development or size. It helps in understanding the regional context of customer behavior. The categories are ordered i.e. Tier 1 > Tier 2 > Tier 3

- DurationOfPitch: Represents the length of the sales pitch made to the customer. Longer durations may indicate more detailed or involved presentations.

- Occupation: Specifies the occupation of the customer, providing insights into the professional background and potentially influencing travel preferences.


- Gender: Indicates the gender of the customer, which can be important for understanding gender-specific preferences in travel.

- NumberOfPersonVisiting: Reflects the number of people visiting, such as family size. This can impact the type and size of travel packages that may be appealing. or simply we can say Total number of persons planning to take the trip with the customer.

- NumberOfFollowups: This is like counting how many times someone tries to talk to you or persuade you after they've initially told you about something, like a travel package.

- ProductPitched: Specifies the type or name of the travel product that was pitched to the customer. Different products may cater to varied preferences and needs.

- PreferredPropertyStar: Indicates the preferred star rating of the accommodation or property. This helps in tailoring offerings to meet specific preferences.

- MaritalStatus: Describes the marital status of the customer, which can influence travel decisions, especially for packages catering to couples or families.

- NumberOfTrips: Represents the number of trips the customer has taken. Previous travel behavior can be indicative of future travel interests.

- Passport: A binary variable indicating whether the customer possesses a passport (1) or not (0). Passport ownership is often a prerequisite for international travel.

- PitchSatisfactionScore: Reflects the customer's satisfaction score related to the sales pitch. This subjective measure gauges the effectiveness of the sales approach.

- OwnCar: A binary variable indicating whether the customer owns a car (1) or not (0). Car ownership may influence travel preferences and logistics.

- NumberOfChildrenVisiting: Indicates the number of children accompanying the customer during the visit, providing insights into family dynamics.

- Designation: Specifies the job designation or title of the customer. Professional status can impact affordability and preferences for travel.

- MonthlyIncome: Represents the monthly income of the customer, providing a key indicator of financial capacity and influencing travel budget considerations.

## Handling Missing values
- For numerical features we will use median to fill the null values and for categorical_features we will fill the null values using the mode.
- Since 'NumberOfFollowups' and 'NumberOfChildrenVisiting' can be treated as discrete numerical features, we'll also use the mode for them.


```python
numerical_columns_with_na=['Age', 'DurationOfPitch', 'NumberOfTrips', 'MonthlyIncome']
categorical_columns_with_na=['TypeofContact', 'PreferredPropertyStar', 'NumberOfFollowups', 'NumberOfChildrenVisiting']
```

```python
for column in numerical_columns_with_na:
    median=data[column].median()
    data[column].fillna(median,inplace=True)
```

```python
# in case there are multiple modes, we will use first mode 

for column in categorical_columns_with_na:
    mode=data[column].mode()[0]
    data[column].fillna(mode,inplace=True)
```

```python
# We will use IQR technique to handle the outliers.
def remove_outlier(data,column):
    Q1=data[column].quantile(0.25)
    Q2=data[column].quantile(0.75)
    IQR=Q2-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q2 + 1.5 * IQR
    new_data=data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return new_data
```

```python
data= remove_outlier(data, 'MonthlyIncome')
```
